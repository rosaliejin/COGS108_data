Our group will focus on finding potential correlations between mental health issues and COVID in the United States.
```python
#Separate the us_search data into a new dataframe for weeks before COVID tracking
us_precovid = us_search[0:32]
```

```python
#Separate the us_search data into a new dataframe for week aftter COVID tracking
us_postcovid = us_search[32:]
```

# Correlations

In order to find the relationship between coronavirus trends and Google search term popularity in the US, we make a condensed dataframe at first for analysis. For our analysis, we're only using data garnered from COVID-19, and excluding pre-COVID data. 

```python
#Define a function to set up the weeks for cases and deaths
def group_week ():
    rows = [0,0,0,0]
    days = 0
    week = 1
    while (days < 126):
        rows.append(week)
        days = days + 1;
        if (days % 7 == 0):
            week = week + 1;
    
    return rows
```

```python
#Add a week column on new_global_cases and sum the data up by week, then return only the cases in the US
new_global_cases = new_global_cases.assign(Week = group_week())
new_global_cases = new_global_cases.groupby("Week").sum()
us_cases = new_global_cases['US']
us_cases
```

```python
#Add a week column on new_global_death and sum the data up by week, then return only the death in the US
new_global_death = new_global_death.assign(Week = group_week())
new_global_death = new_global_death.groupby("Week").sum()
us_death = new_global_death['US']
us_death
```

```python
#Add the values of cases into us_postcovid dataset
us_postcovid = us_postcovid.reset_index(drop = True)
combined_data =pd.concat([us_postcovid, us_cases], axis=1)
combined_data
```

```python
#Rename the column
combined_data = combined_data.rename(columns={'US': 'Cases'})
```

```python
#Same operation on death
combined_data =pd.concat([combined_data, us_death], axis=1)
combined_data = combined_data.rename(columns={'US': 'Death'})
combined_data
```

```python
# import the packages needed for the analysis
import patsy
import statsmodels.api as sm
from scipy.stats import pearsonr, norm, ttest_ind
```

Now we are ready to explore the correlations between different search terms and COVID confirmed cases and deaths in our dataset.

```python
combined_data.corr()
```

In this correlation analysis, we can find that the terms 'depression', 'obsessive compulsive disorder', 'ocd', 'panic attack' and 'mental health' have little correlation (r less than positive or negative 0.4) with both confirmed cases and death cases. However, the term 'insomnia' has strong positive correlation to both confirmed cases and death cases while the terms 'anxiety', 'counseling' and 'psychiatrist' have strong negative correlation to both confirmed cases and death cases. Overall, 'counseling' has the largest negative Pearson's correlation while 'insomnia' has the largest positive Pearson's correlation.

Since 'counseling' and 'psychiatrist' have the most negative Pearson's corelations according to our analysis, let's take a deeper look at them by building an OLS models to further investigate the relationships.

Here when we build our OLS models, we determined that our cutoff for the p-value should be less than 0.04. In this case, our null hypothesis is that there is no correlation between COVID confirmed cases/deaths and the mental health search terms we are trying to measure. Our alternative hypothesis therefore becomes that there is a statistically significant correlation between COVID confirmed cases/deaths and the mental health search terms.


```python
#generate linear regession model for the data of counseling search term and confirmed cases
outcome, predictors = patsy.dmatrices('counseling ~ Cases',combined_data)
mod = sm.OLS(outcome, predictors)
#fit the model
res = mod.fit() 
#print the results
print(res.summary())
```

Note: the reason that the coefficient is so small is mainly because of the scaling of our data is so different where the COVID cases go up to 200000 eventually. The condition number is large for the same reason. However, the p-value is small for our analysis, which indicate the linear relationship we are finding with these two variables are valid.

```python
#generate scatter plot for the data of counseling search term and confirmed cases
sns.scatterplot(x='Cases', y='counseling', data=combined_data, alpha=0.3, s=100).set_title('Search Score for Counseling vs. Total Confirmed Cases')
#plot of the linear regression model generated
xs = np.arange(combined_data['Cases'].min(), combined_data['Cases'].max())
ys = -0.0001 * xs + 87.6484
plt.plot(xs, ys, '--k', linewidth=4, label='Model')

plt.ylabel('Search Score for Counseling')
plt.xlabel('Total Confirmed Cases')
plt.legend();
```

```python
#generate linear regession model for the data of counseling search term and death cases
outcome, predictors = patsy.dmatrices('counseling ~ Death',combined_data)
mod = sm.OLS(outcome, predictors)
#fit the model
res = mod.fit() 
#print the results
print(res.summary())
```

#Ethics and Privacy
The question we are seeking to answer has ties to mental health, which is a topic that is very private and personal to individuals. As such, for data collection, when finding data related to this topic it must be anonymous and should not breach the privacy of any individual. It also must be data given under informed consent. Another thing we had to keep in mind is that there could easily be a lot of bias when it comes to mental health data. For example, certain groups of people may be facing proportionally higher stress factors than other groups, such as individuals who are socioeconomically disadvantaged and may be having a harder time during the pandemic. Hence, the sample must be diverse and not focus on any particular group. 

To address these concerns, we are observing mental health through search term datasets on Kaggle that were collected from Google Trends. The advantage of data from Trends is that it is open-source, anonymized and representative of all Google searches in a particular region (Source: [Medium](https://medium.com/google-news-lab/what-is-google-trends-data-and-what-does-it-mean-b48f07342ee8)). Being anonymized, Google Trend data does not tell us who is searching online about mental health and thus privacy is maintained for individuals. Also, Google Trends data is collected under informed consent, as it is noted explicitly in their Privacy Policy: “we share information publicly to show trends”. (Source: [Google](https://policies.google.com/privacy?hl=en-US#infosharing)). Hence, the data for mental health we use does not violate any privacy considerations. In terms of bias, Google Trends data is for all Google searches in a region, so it is not excluding data based on any factor but geography. Also, since we are pulling search data for entire countries, we minimize any location biases (i.e. if data was looked at for a specific city that is poorer or one that is richer) and can say the search data is representative of a country as a whole. This helps address concerns about biases as the data considers everyone who uses Google fairly. However, there is an inherent bias that we could not completely eliminate and it is that since Google Search is a web-based service, our sample is people with internet access. Any conclusions we draw cannot be generalized to individuals without internet access. 

The other dataset utilized in this project is COVID-19 cases and deaths data. Since such data is related to medical health and typically country specific, where the data comes from and its usage rights is important. Also, it needs to be representative of a country. For this project, we used a dataset on Kaggle that has data from the John Hopkins University Coronavirus Global Dashboard. JHU uses publicly available data to compile this dashboard, so privacy concerns are minimized. Also, there is no way to tie a particular case or death data point to any one individual, so it is anonymized. In terms of bias, one thing we do want to watch out for is the regional differences within a country, such as differences in coronavirus impact depending on the wealth of a particular city. To get a representative figure for a country, we consider the countries as a whole, taking into account all of the regions as one. Another potential source of bias is how much testing/reporting is being done as a whole for a country. As such, in our analysis we do not weigh countries against each other, such as saying Country A is more impacted than B, but instead weight things like Country A at time X and Country A at time Y. Additionally, we do correlational studies focusing on the United States, which has fairly robust and regular COVID testing. Pervasiveness of coronavirus testing in the U.S. minimizes potential bias from undiscovered or unreported COVID cases. In all, the COVID data used passes privacy considerations and we utilize it in a way to minimize bias. 

Overall, the data we use for this project has been collected fairly and passes our privacy considerations. They are not free of bias, but we minimize the effects of it with thoughtful analysis. In terms of intent, the purpose of this project is to discover possible correlations between COVID-19 trends and mental health, without any goal to implicate or harm any individuals. Due to the fact that our data is anonymized, this project cannot be used to single out or direct harm to an individual. It is also not meant to lend any meaning to how X Country is handling the pandemic and cannot be used to that end as measures by countries are not analyzed. At its base, this project just offers an unbiased study on the possible correlation COVID-19 trends and mental health.
